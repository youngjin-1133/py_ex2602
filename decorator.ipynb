```python
# 클로저 : 안쪽 함수가 바깥 변수 값을 기억하는 구조

def outer():
    def inner():
        print('안쪽 함수')
    return inner


def make_plus(x):
    def plus(n):
        return x + n
    return plus

rs = make_plus(3)

print(rs(5))
print(rs(10))
```

```python
# 데코레이터 : 기존 함수를 고치지 않고, 기능만 덧붙이는 방법

def deco(fun):
    def wrap():
        print('실행 전')
        fun()
        print('실행 후')
    return wrap

# def hello():
#     print('안녕하세요')

# new_fun = deco(hello)
# new_fun()

@deco
def hello():
    print('안녕하세요')

hello()
```
