```python
# 정규표현식(정규식)

text = "안녕하세요 저는 15살 입니다."

print("15" in text)
```

```python
import re

text = "안녕하세요"

# \d는 숫자
# +는 1개 이상
pattern = r"\d+" # r은 raw string \(특수문자)
result = re.findall(pattern, text)
print(result)
```

```python
# 패턴 찾기
sample = "abc123456"

re.findall(r"\d+", sample)

re.findall(r"[a-z]+", sample)
```

```python
# 특정 형식 찾기

data = "test123@gmail.com"

email_patttern = r"\w+@\w+.\w+"
# \w는 문자 또는 숫자
# + 1개 이상
# @ 말 그대로 @
# \. : 점(특수 문자이기 때문에 \를 붙인다.)

emails = re.findall(email_patttern, data)
print(emails)
```
