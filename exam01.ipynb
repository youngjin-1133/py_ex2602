```python
# 문제 : 실행되는 출력문에는 `참` 그렇지 않으면 `거짓` 이라고 적어주세요.
# 조건 : `참1`, `참2`, `거짓1` 이런 순으로 적어주세요.

# 개념시작
if ( True ):
  print("참")
    
if ( False ):
  print("거짓")
# 개념끝

# 문제시작
a = 10

# `==` => 같다.
if ( a == 10 ):
  print("참1")

# `!=` => 같지 않다.
if ( a != 10 ):
  print("거짓1")

if ( a > 10 ):
  print("거짓2")

if ( a >= 10 ):
  print("참2")

b = 10

if ( a == b ):
  print("참3")

# c 에는 true or false 이 담긴다.
c = a != b

if ( c ):
  print("거짓3")

if ( c == False ):
  print("참4")

# `not` => 반전
if ( not c ):
  print("참5")

if ( not(not c) ):
  print("거짓4")

d = True

if ( c != d ):
  print("참6")

if ( 10 > 5 and -6 > -10 and 10 != 10 ):
  print("거짓5")

if ( 10 < 5 and -6 < -10 and 10 != 10 or 10 > 5 ):
  print("참7")
# 문제끝
```
