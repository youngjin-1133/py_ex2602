```python
import requests as req
from bs4 import BeautifulSoup

response = req.get("https://news.naver.com/")
#print(response)
#print(response.text)

# html 문자열 파싱
soup = BeautifulSoup(response.text, 'html.parser')

content = soup.select('.cnf_news_title')
print(content[0].text)

print('== 반복문으로 텍스트 출력 ==')
for c in content:
    print(c.text)

content2 = soup.select('.cn_title.area')

print('== 자식 선택자로 선택, 반복문으로 텍스트 출력 ==')
for c in content2:
    print(c.text)

imgs = soup.select('.cnf_news_thumb img')

print("== 자식 선택자로 선택, 반복문으로 이미지 주소 출력 ==")
for img in imgs:
    print(img) # print(img['src'])
```
