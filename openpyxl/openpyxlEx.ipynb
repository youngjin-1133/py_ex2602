```python
from openpyxl import *

# ==== 딕셔너리로 값 생성 후 엑셀에 넣기 ====

item1 = {
    '품목': '바나나',
    '가격': '30000',
    '수량': 1,
    '날짜': '2026년 2월 21일'
}

item2 = {
    '품목': '사과',
    '가격': '16000',
    '수량': 2,
    '날짜': '2026년 2월 22일'
}

item3 = {
    '품목': '오렌지',
    '가격': '25000',
    '수량': 4,
    '날짜': '2026년 2월 23일'
}

item4 = {
    '품목': '메론',
    '가격': '50000',
    '수량': 1,
    '날짜': '2026년 2월 24일'
}

wb = Workbook()
ws = wb.active

ws['A1'] = '품목' # A열 1행 : 품목
ws['B1'] = '가격' # A열 1행 : 품목
ws['C1'] = '수량' # A열 1행 : 품목
ws['D1'] = '날짜' # A열 1행 : 품목

items = [item1, item2, item3, item4]

i = 2
for item in items:
    ws['A' + str(i)] = item['품목']
    ws['B' + str(i)] = item['가격']
    ws['C' + str(i)] = item['수량']
    ws['D' + str(i)] = item['날짜']
```
