```python
# openpyxl => 파이썬에서 엑셀 파일을 읽고 쓰기 위한 라이브러리

# openpyxl 불러오기 / 모든 기능을 불러옴
# import openpyxl
from openpyxl import *

# 버전 확인
# print(openpyxl.__version__)


# ==== 엑셀 파일 생성, 기본 시트 선택, 셀 값 넣기 ====

# 새로운 엑셀 파일 만들기
# 새 엑셀 파일을 만들겠다 라는 의미
w1 = Workbook()

# 만들어진 엑셀 파일에서 기본 시트를 선택
ws = w1.active

# 지금 만든 엑셀 파일을 실제 파일로 저장
# 파일의 이름은 test.xlsx로 저장
# w1.save('test.xlsx')

# 기존에 저장된 엑셀 파일을 다시 열어서 수정할 수 있게 함
w2 = load_workbook('test.xlsx')

# 열어온 엑셀 파일에서 첫 번째 시트 선택
ws2 = w2.active

# 셀에 값 입력하기
# A열 1행 셀에 hello라는 텍스트 입력
ws2['A1'] = 'hello'

# 변경사항 저장
w2.save('test.xlsx')

# C열 3행 셀에 bye라는 텍스트 입력
ws2['C3'] = 'bye'

# 변경사항 저장
w2.save('test.xlsx')

a = 'python'
# ws2['D5':'F8'] : D5부터 F8셀까지의 범위를 선택(4행 3열 = 총 12개의 셀)
for row in ws2['D5':'F8']: # 행 단위로 반복
    for cell in row: # 각 행 안의 셀들을 하나씩 꺼내서
        cell.value = a # 셀엘 python이라는 값 입력

# 변경사항 저장
w2.save('test.xlsx')

num = 2
# A1부터 C3까지 범위 선택(3행 * 3열 = 9개 셀)
for row in ws2['A1':'C3']:
    for cell in row:
        cell.value = num
        num += 2

# 변경사항 저장
w2.save('test.xlsx')
```

```python
from openpyxl import *

# ==== 딕셔너리로 값 생성 후 엑셀에 넣기 ====

item1 = {
    '품목': '바나나',
    '가격': '3000',
    '수량': 1,
    '날짜': '2026년 2월 20일'
}

item2 = {
    '품목': '책',
    '가격': '15000',
    '수량': 3,
    '날짜': '2026년 2월 22일'
}

item3 = {
    '품목': '커피',
    '가격': '4000',
    '수량': 10,
    '날짜': '2026년 2월 23일'
}

item4 = {
    '품목': '신발',
    '가격': '50000',
    '수량': 4,
    '날짜': '2026년 2월 26일'
}

wb = Workbook()
ws = wb.active

ws['A1'] = '품목' # A열 1행 : 품목
ws['B1'] = '가격' # B열 1행 : 품목
ws['C1'] = '수량' # C열 1행 : 품목
ws['D1'] = '날짜' # D열 1행 : 품목

items = [item1, item2, item3, item4]

i = 2
for item in items:
    ws['A' + str(i)] = item['품목']
    ws['B' + str(i)] = item['가격']
    ws['C' + str(i)] = item['수량']
    ws['D' + str(i)] = item['날짜']
    i += 1

wb.save('test2.xlsx')
```
