```python
fig = plt.figure()

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3, 4] # 막대의 위치(좌표 기준)
y = [1, 2] # 막대의 높이(값 기준)

# x축 라벨 이름(tick 이름)
labels = ['english', 'math']

# 각 막대에 적용할 색상
colors = ['red', 'blue']

a.bar(x, y, tick_label=labels, color=colors) # bar 앞에 h를 붙이면 가로로 나옴
plt.show()
```

```python
fig = plt.figure()

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3, 4] # 막대의 위치(좌표 기준)
y = [1, 2] # 막대의 높이(값 기준)

# x축 라벨 이름(tick 이름)
labels = ['english', 'math']

# 각 막대에 적용할 색상
colors = ['red', 'blue']

a.bar(x, y, tick_label=labels, color=colors) # bar 앞에 h를 붙이면 가로로 나옴

for i, v in enumerate(y):
    a.text(v+2, i + 1.05, str(v)) # 표준 수치
plt.show()
```
