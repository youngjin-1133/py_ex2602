```python
# 클래스 외부에서는 접근 불가, 클래스 내부에서만 사용 가능한 private attribute
class Person():
    def __init__(self, name, age, hobby, weight):
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.hobby = hobby
        self.__weight = weight

    def diet(self, loss_weight):
        self.__weight -= loss_weight
        print("몸무게 :", self.__weight)

jin = Person('진', 22, '운동', 78)
jin.diet(10)
```

```python
# 클래스 속성

class person():
    folder = [] # 클래스 속성 -> 클래스에 속한다(공공제) 모든 인스턴스에서 접근이 가능하다

    def add files(self, file):
        self. folder.append(file) # 클래스 이름으로 클래스 속성에 접근할 수 있다
    
    jin = person()
    jin.add_files('txt')

    paul = person()
    paul.add_files('img')

    print(jin.folder)
    print(paul.folder)
```

```python
# 인스턴스 속성
class Person():
    def __init__(self):
        self.folder = [] # 인스턴스 속성

    def add_files(self, file):
        self.folder.append(file) # 클래스 이름으로 클래스 속성에 접근할 수 있다.

jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')

print(jin.folder)
print(paul.folder)
```

```python
# 비공개 클래스 속성
class Person:
    __weight = 78

    
    def print_weight(self):
        print(Person.__weight)

jin = Person()
jin.print_weight()

print(jin.__weight)
```
